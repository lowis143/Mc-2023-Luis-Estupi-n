```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style


plt.rcParams['image.cmap'] = "bwr"
#plt.rcParams['figure.dpi'] = "100"
plt.rcParams['savefig.bbox'] = "tight"
style.use('ggplot') or plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')


#datos



ejex = [2,4,6,8,10,12]
ejey = [2.2,3,4.5,6,8.5,12]

#ejex = [1,2,3,4,5]
#ejey = [5,6.6,8.5,10.8,14.2]

datos = pd.DataFrame({'X': ejex, 'Y': ejey})
datos.head(6)
```

```python
m, b = np.polyfit(ejex, ejey, 1)
x_line = np.linspace(0, 12, 100)
y_line = m * x_line + b
plt.scatter(ejex, ejey)
plt.plot(x_line, y_line, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión lineal por mínimos cuadrados')
plt.show()
```

```python
plt.plot(ejex, ejey);
plt.show ()
```

```python
# Calculamos la pendiente y la ordenada al origen por mínimos cuadrados
m, b = np.polyfit(ejex, ejey, 1);

# Imprimimos los resultados
print(f"La ecuación de la recta de mínimos cuadrados es y = {b}x + {m}")
```

reemplazamos log natural de Y  por Y 

```python
#ejex = [2,4,6,8,10,12]
ejey1 = []
for y in ejey:
  ejey1.append(math.log(y))

datos = pd.DataFrame({'X': ejex, 'Y': ejey1})
datos
```

```python
m, b = np.polyfit(ejex, ejey1, 1)
x_line = np.linspace(0, 12, 100)
y_line = m * x_line + b
plt.scatter(ejex, ejey1)
plt.plot(x_line, y_line, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión lineal por mínimos cuadrados')
plt.show()
```

```python
# Calculamos la pendiente y la ordenada al origen por mínimos cuadrados
m, b = np.polyfit(ejex, ejey1, 1);

# Imprimimos los resultados
print(f"La ecuación de la recta de mínimos cuadrados es y = {b} + {m}x")
```

```python
alfa = math.exp(b)

# Imprimir el resultado
print(alfa)
```

```python
#m, b = np.polyfit(ejex, ejey1, 1)
x_line = np.linspace(0, 12, 100)
y_line = alfa * math.e**(m*x_line)
plt.scatter(ejex, ejey)
plt.plot(x_line, y_line, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Modelo exponencial')
plt.show()
```
