```python
#primera gr치fica
X = [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5]
Y = [4.2,6.5,7.5,8,8.5,8.8,9,9.1]


promx = (sum(X)/8)
print("promedio de X es de:",promx)

promy = (sum(Y)/8)
print("promedio de Y es de:",promy)


xy = [X[i] * Y[i]
      for i in range(len(X))]
                  

print("sumatoria de x*y:",sum(xy))

x2 = [X[i]**2
      for i in range(len(X))]
print("sumatoria de x^2:", sum(x2))


a1 = (len(X)*sum(xy)-sum(X)*sum(Y))/(len(X)*sum(x2)-(sum(X)**2))
print("a1 es igual a:",a1)

ao = promy - (a1*4)
print("ao es igual a:",ao)

#Segunda gr치fica
import math 
X = [0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5]
Y = [1.435084525,1.871802177,2.014903021,2.079441542,2.140066163,2.174751721,2.197224577,2.208274414]

promx = (sum(X)/8)
print("promedio de X es de:",promx)

promy = (sum(Y)/8)
print("promedio de Y es de:",promy)


xy = [X[i] * Y[i]
      for i in range(len(X))]
                  

print("sumatoria de x*y:",sum(xy))

x2 = [X[i]**2
      for i in range(len(X))]
print("sumatoria de x^2:", sum(x2))


a1 = (len(X)*sum(xy)-sum(X)*sum(Y))/(len(X)*sum(x2)-(sum(X)**2))
print("a1 es igual a:",a1)

ao = promy - (a1*promx)
print("ao es igual a:",ao)
print("--------------------------------")


alpha = math.e**ao
beta = a1

print("alpha igual a:",alpha)
print("beta igual a:",beta)

#Tercera gr치fica
import math 
X = [-0.301029996,0.176091259,0.397940009,0.544068044,0.653212514,0.740362689,0.812913357,0.875061263]

Y = [0.62324929,0.812913357,0.875061263,0.903089987,0.929418926,0.944482672,0.954242509,0.959041392]

promx = (sum(X)/8)
print("promedio de X es de:",promx)

promy = (sum(Y)/8)
print("promedio de Y es de:",promy)


xy = [X[i] * Y[i]
      for i in range(len(X))]
                  

print("sumatoria de x*y:",sum(xy))

x2 = [X[i]**2
      for i in range(len(X))]
print("sumatoria de x^2:", sum(x2))


a1 = (len(X)*sum(xy)-sum(X)*sum(Y))/(len(X)*sum(x2)-(sum(X)**2))
print("a1 es igual a:",a1)

ao = promy - (a1*promx)
print("ao es igual a:",ao)
print("--------------------------------")


alpha = 10**ao
beta = a1

print("alpha igual a:",alpha)
print("beta igual a:",beta)

#cuarta gr치fica
import math 
X = [2,0.666666667,0.4,0.285714286,0.222222222,0.181818182,0.153846154,0.133333333]
Y = [0.238095238,0.153846154,0.133333333,0.125,0.117647059,0.113636364,0.111111111,0.10989011]

promx = (sum(X)/8)
print("promedio de X es de:",promx)

promy = (sum(Y)/8)
print("promedio de Y es de:",promy)


xy = [X[i] * Y[i]
      for i in range(len(X))]
                  

print("sumatoria de x*y:",sum(xy))

x2 = [X[i]**2
      for i in range(len(X))]
print("sumatoria de x^2:", sum(x2))


a1 = (len(X)*sum(xy)-sum(X)*sum(Y))/(len(X)*sum(x2)-(sum(X)**2))
print("a1 es igual a:",a1)

ao = promy - (a1*promx)
print("ao es igual a:",ao)
print("--------------------------------")


alpha = 1/ao 
beta = a1/ao

print("alpha igual a:",alpha)
print("beta igual a:",beta)
```
