```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada
x = [1, 3, 5, 7, 9]
fx = [2.8, 3, 3.5, 3.8, 4.5]
value_to_estimate = 4.75

# Grados de los polinomios de interpolación
degrees = [1, 2, 3, 4]

# Función para calcular los polinomios de interpolación de Lagrange
def lagrange_interpolation(x, fx, degree, value):
    n = len(x)
    result = 0.0

    for i in range(n):
        # Calcular el polinomio de interpolación de Lagrange
        term = fx[i]
        for j in range(n):
            if i != j:
                term *= (value - x[j]) / (x[i] - x[j])
        result += term

    return result

# Calcular los polinomios de interpolación y estimar f(4.75) para cada grado
estimations = []
for degree in degrees:
    estimation = lagrange_interpolation(x, fx, degree, value_to_estimate)
    estimations.append(estimation)

    # Gráfica del polinomio de interpolación
    plt.figure()
    x_interp = np.linspace(min(x), max(x), 100)
    y_interp = lagrange_interpolation(x, fx, degree, x_interp)
    plt.plot(x_interp, y_interp, label=f'Grado {degree}')
    plt.scatter(x, fx, color='red', label='Datos')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()

plt.show()
```

```python
# Estimación de f(4.75) para cada grado
for i, degree in enumerate(degrees):
    print(f'Estimación (grado {degree}): {estimations[i]}')
```

```python
def lagrange_interp(x, fx, degree, point):

    # Calcular los coeficientes de los polinomios de interpolación de Lagrange.
    coeffs = []
    for j in range(degree + 1):
        term = fx[j]
        for k in range(degree + 1):
            if k != j:
                term *= (point - x[k]) / (x[j] - x[k])
        coeffs.append(term)

    # Sumar los términos del polinomio.
    result = sum(coeffs)

    return result


# Puntos base
x = [1, 3, 5, 7, 9]
fx = [2.8, 3, 3.5, 3.8, 4.5]

# Polinomios de interpolación de Lagrange de grado 1, 2 y 3
L1 = lagrange_interp(x, fx, 1, 4.75)
L2 = lagrange_interp(x, fx, 2, 4.75)
L3 = lagrange_interp(x, fx, 3, 4.75)

# Polinomios de grado 1 y 2 asociados a los puntos seleccionados
L1_points = lagrange_interp(x[:2], fx[:2], 1, 4.75)
L2_points = lagrange_interp(x[:3], fx[:3], 2, 4.75)

# Imprimir los resultados
print("Polinomio de interpolación de Lagrange de grado 1: ", L1)
print("Polinomio de interpolación de Lagrange de grado 2: ", L2)
print("Polinomio de interpolación de Lagrange de grado 3: ", L3)
print("Polinomio de grado 1 asociado a los puntos seleccionados: ", L1_points)
print("Polinomio de grado 2 asociado a los puntos seleccionados: ", L2_points)
```
