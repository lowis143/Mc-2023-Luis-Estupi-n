```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos de entrada
ejex = [1, 2, 3, 4, 5, 6, 7, 8]
ejey = [3.8, 4, 4.2, 4.6, 4.9, 5.3, 5.7, 6.1]

# Función exponencial
def exponential_func(x, a, b, c):
    return a * np.exp(b * x) + c

# Función de potencias
def power_func(x, a, b, c):
    return a * np.power(x, b) + c

# Ajuste exponencial
popt_exp, pcov_exp = curve_fit(exponential_func, ejex, ejey)

# Ajuste de potencias
popt_pow, pcov_pow = curve_fit(power_func, ejex, ejey)

# Razón de crecimiento
slope = (ejey[-1] - ejey[0]) / (ejex[-1] - ejex[0])

# Gráfica de los puntos originales
plt.scatter(ejex, ejey, label='Datos')

# Gráfica de la función exponencial
x_exp = np.linspace(min(ejex), max(ejex), 100)
y_exp = exponential_func(x_exp, *popt_exp)
plt.plot(x_exp, y_exp, 'r-', label='Exponencial')


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Gráfica de los puntos originales
plt.scatter(ejex, ejey, label='Datos')
# Gráfica de la función de potencias
x_pow = np.linspace(min(ejex), max(ejex), 100)
y_pow = power_func(x_pow, *popt_pow)
plt.plot(x_pow, y_pow, 'g--', label='Potencias')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Gráfica de los puntos originales
plt.scatter(ejex, ejey, label='Datos')
# Gráfica de la razón de crecimiento
x_growth = np.linspace(min(ejex), max(ejex), 100)
y_growth = ejey[0] + slope * (x_growth - ejex[0])
plt.plot(x_growth, y_growth, 'b:', label='Crecimiento')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
